```
import sympy as sp
import sympy.plotting as spl
```

```
x = sp.symbols('x')
sp.init_printing()
```

---

$$ \Large 
f(x) = 3 \sin{\left(\frac{x}{3} \right)} \cos{\left(3 x \right)}
$$
$$\Large 
{\displaystyle g(x)=\frac{5}{9}x^{3} + \frac{5}{3}x^{2} + \frac{1}{9}x - \frac{2}{3}\pi}
$$

What is the value of:

$$ \Large
{\frac {d}{dx}\, \left( f(x)\cdot g(x) \right)   }\ \ \mathrm{at} {\ \ x\, =\, 3\pi}
$$
Don't use decimal approximations for the rational numbers.
I want an exact answer, not a decimal approximation.

```
my_fx=3 * sp.sin(x/3) * sp.cos(3*x)
```

```
my_gx = sp.Rational(5, 9)*x**3 + sp.Rational(5, 3)*x**2 + sp.Rational(1, 9)*x - sp.Rational(2, 3)*sp.pi
```

```
my_equ = sp.diff(my_fx*my_gx, x)

result = sp.simplify(my_equ.subs(x, 3*sp.pi))

sp.expand(result)
```

Plot $f(x)$ and $g(x)$ on the same plot over the interval $[-4 \le x \le 2]$

```
spl.plot(my_fx, my_gx,
        (x, -4.0, 2.0),
        title = "f(x) & g(x) over the [-4,2]", 
        xlabel = 'x', 
        ylabel = 'my function',
        xlim = (-4.0, 2.0),
        ylim = (-7,7),
        size = (8, 6),
        legend = True,
        axis_center = (0, 0)
       );
```

What are the value(s) for $x$ where $f(x)\ = \ g(x)$ over the interval $[-4 \le x \le 2]$

```
all_guesses = (-2.5, -1.8, 0.8)

for val in all_guesses:
    result = sp.nsolve(my_fx - my_gx, x, val)
    print(result)
```
