HQ Gaia Data - Galactic Stellar Populations

```
import numpy as np

from astropy.table import QTable
from astroquery.gaia import Gaia
```

Stellar population at high galactic latitudes

Create and run a Gaia query to find a
high quality complete representitive sample
of at least 2,000 objects at a high galactic latitude (l = 45°, b = 80°) that are brighter than 17th magnutude, have a parallax between 1.0 and 2.0, have a parallax SNR > 10, and have the bp-rp colors.
You will want the columns:
source_id, l, b, phot_g_mean_mag, bp_rp, parallax, parallax_over_error
Make sure the number of objects is less than the query limits
Don't print out the entire 2,000 object table!

```
my_query = """
SELECT TOP 2500
source_id,l,b,phot_g_mean_mag,bp_rp,parallax,parallax_over_error
FROM gaiadr3.gaia_source_lite
WHERE DISTANCE( POINT(45, 80), POINT(l, b) ) < 2.5
AND phot_g_mean_mag < 17
AND parallax BETWEEN 1.0 AND 2.0
AND parallax_over_error > 10
AND bp_rp IS NOT NULL
"""
```

```
my_job_query = Gaia.launch_job(my_query)
```

```
print(my_job_query)
```

```
target_table = my_job_query.get_results()
target_table[0:5]
```

Use
parallax
and
phot_g_mean_mag
to get
distance
(in pc) and the
absolute magnitude
($G_{M}$)

Add them as columns to the data table

```
def find_distance(my_parallax):
    result = 1 / (my_parallax / 1000)
    return result
    
def find_absmag(my_appmag, my_distance):
    result = my_appmag - (5 * np.log10(my_distance / 10))
    return result
```

```
target_table['distance'] = find_distance(
                               my_parallax = target_table['parallax']
                           )

target_table['Abs_G_Mag'] = find_absmag(
                               my_appmag = target_table['phot_g_mean_mag'],
                               my_distance = target_table['distance']
                         )
```

```
target_table[0:5]
```

Create a subset of Red giant stars

An Absolute G magnitude
brighter
than 4th magnitude
BP-RP
color index
greater
than 1.0
You do not need to a do a new Gaia search, just filter/mask the data from the table you already made.

```
giant_table=target_table[(target_table['bp_rp'] > 1.0)
    &(target_table['Abs_G_Mag'] < 4.0)]
giant_table
```

Create a subset of Young Blue stars

An Absolute G magnitude
brighter
than 4th magnitude
BP-RP
color index
less
than 1.0
You do not need to a do a new Gaia search, just filter/mask the data from the table you already made.

```
young_table=target_table[(target_table['bp_rp'] < 1.0)
    &(target_table['Abs_G_Mag'] < 4.0)]
young_table
```

What is the ratio of the number Young Blue stars/Red giant stars for your HIGH galactic latitude field.

```
np.size(young_table) / np.size(giant_table)
```
