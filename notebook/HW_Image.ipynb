Get the World Coordinate System (WCS) data from your FITS file.

```
target_data   = fits.getdata(target_file, 0)
target_header = fits.getheader(target_file, 0)

target_wcs = WCS(target_header)
```

```
fig, ax = plt.subplots(
    figsize=(8, 5),
    constrained_layout=True)

fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("A histogram of the pixel values of the target image")
ax.set_xlabel("Image Pixel Values")
ax.set_ylabel("Number of Pixels")
ax.hist(target_data.flatten(),bins=100);
```

```
fig, ax = plt.subplots(
        subplot_kw={'projection': target_wcs},
    figsize=(8, 5), constrained_layout=True)


fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("Target Image")
ax.imshow(target_data, 
               cmap = plt.cm.gray,
                origin='lower')
```

```
target_data[target_data < target_data.mean()] = target_data.mean()


fig, ax = plt.subplots(
    subplot_kw={'projection': target_wcs},
    figsize=(8, 5), constrained_layout=True)


fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("Target Image where (values < MEAN set to min)")
ax.imshow(target_data, 
               cmap = plt.cm.gray,
                origin='lower')
```

```
target_data[target_data < 0.6 * target_data.max()] = target_data.min()


fig, ax = plt.subplots(
    subplot_kw={'projection': target_wcs},
    figsize=(8, 5), constrained_layout=True)


fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("Target image where (values < (60% of MAX) set to min)")
ax.imshow(target_data, 
               cmap = plt.cm.gray,
                origin='lower')
```

```
target_data[target_data < 0.9 * target_data.max()] = target_data.min()


fig, ax = plt.subplots(
    subplot_kw={'projection': target_wcs},
    figsize=(8, 5), constrained_layout=True)


fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("Target image where (values < (90% of MAX) set to min)")
ax.imshow(target_data, 
               cmap = plt.cm.gray,
                origin='lower')
```
