The Solar Neighborhood

The Solar Neighborhood can be roughly defined as all of the objects within 10 pc of the Sun.
Objects within 10 pc of the Sun will have parallax values greater than 100 mas.
Create and run a Gaia query to find a high quality complete sample of all of the objects Gaia surveyed within the Solar Neighborhood.
Your sample should only have high quality data (parallax SNR > 15)
All of your objects should have bp-rp colors and radial velocity measurements.
You will want the columns:
source_id, bp_rp, parallax, parallax_over_error, pmra, pmdec, radial_velocity
Don't print out the entire object table!

```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
import astropy.units as u

from astroquery.simbad import Simbad

from astropy.table import QTable
from astroquery.gaia import Gaia
```

```
my_query = f"""
SELECT TOP 500
source_id, bp_rp, parallax, parallax_over_error, pmra, pmdec, radial_velocity
FROM gaiadr3.gaia_source_lite
WHERE parallax > 100
AND bp_rp IS NOT NULL
AND radial_velocity IS NOT NULL
ORDER BY parallax DESC
"""
```

```
my_job_query = Gaia.launch_job_async(my_query)
```

```
print(my_job_query)
```

```
my_table = my_job_query.get_results()
my_table[0:5]
```

Calculate the total space velocity for the Solar Neighborhood objects

```
my_table['Total_PM'] = np.sqrt(my_table['pmra'] ** 2 + my_table['pmdec'] ** 2)
```

```
my_table['Tan_Velocity'] = 4.74047 * ( my_table['Total_PM'] / my_table['parallax'] )
```

```
my_table['Space_Velocity'] = np.sqrt(my_table['Tan_Velocity'] ** 2 + my_table['radial_velocity'] ** 2)
```
