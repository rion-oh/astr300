```
import matplotlib.pyplot as plt

import numpy as np
import astropy.units as u
from astropy.time import Time
import pytz

from astroplan.plots import plot_airmass
from astroplan import Observer, FixedTarget, time_grid_from_range

import warnings
warnings.simplefilter('ignore', Warning)
```

```
# Location 1

my_location_one = Observer(longitude = -73.14074750668829 * u.deg,
                         latitude = -48.55474333867469 * u.deg,
                         timezone = 'America/Santiago',
                         name = "Cerro Mellizo Sur Aysén, Chile"
                        )
```

```
# Location 2

my_location_two = Observer(longitude = 117.99135843232759 * u.deg,
                         latitude = -8.24720230309624 * u.deg,
                         timezone = 'Asia/Makassar',
                         name = "Mount Tambora West Nusa Tenggara, Indonesia"
                        )
```

```
# Location 3

my_location_three = Observer(longitude = -62.34584445101083 * u.deg,
                         latitude = 82.49886570465698 * u.deg,
                         timezone = 'America/Toronto',
                         name = "Canadian Forces Station Alert Nunavut, Canada"
                        )
```

Determine the UTC time for
Local Noon
at each of the locations

Observing Date:
Dec 24, 2026

You will have to experiment with the
which =
parameter
Make sure local noon occurs on the observing date

```
# Location 1

my_date = Time("2026-12-24") 
my_noon_one = my_location_one.noon(my_date, which="next")

print(my_noon_one.iso)
```

```
# Location 2

my_noon_two = my_location_two.noon(my_date, which="next")

print(my_noon_two.iso)
```

```
# Location 3

my_noon_three = my_location_three.noon(my_date, which="next")

print(my_noon_three.iso)
```

Determine the length of the observing run at each location

Length of run is defined by
Astronomical Twilight
If your length is negative (or > 24 hr), you have to use a different
which =
parameter above
Your output should look like -
At LOCATION the total observing time is TIME hours
The observing length should be VERY different at each location

```
# Location 1

observe_night_one = my_location_one.tonight(my_noon_one, horizon = -18 * u.deg)
```

```
# Print out observing length

observing_length_one = (observe_night_one[1] - observe_night_one[0]).to(u.h)

print(f"At {my_location_one.name} the total observing time is {observing_length_one.value:.1f} hours")
```

```
# Location 2

observe_night_two = my_location_two.tonight(my_noon_two, horizon = -18 * u.deg)
```

```
# Print out observing length

observing_length_two = (observe_night_two[1] - observe_night_two[0]).to(u.h)

print(f"At {my_location_two.name} the total observing time is {observing_length_two.value:.1f} hours")
```

```
# Location 3

observe_night_three = my_location_three.tonight(my_noon_three, horizon = -18 * u.deg)
```

```
# Print out observing length

observing_length_three = (observe_night_three[1] - observe_night_three[0]).to(u.h)

print(f"At {my_location_three.name} the total observing time is {observing_length_three.value:.1f} hours")
```

Set up the Target list

NGC 6720

```
target1 = FixedTarget.from_name("NGC 6720")
target2 = FixedTarget.from_name("NGC 2548")
target3 = FixedTarget.from_name("NGC 2068")
target4 = FixedTarget.from_name("NGC 5024")
target5 = FixedTarget.from_name("NGC 2535")
target6 = FixedTarget.from_name("NGC 2685")
```

```
target_list = [target1, target2, target3, target4, target5, target6]
target_list
```

What can you see at each location?

For each location, plot airmass vs. time for all of the targets over the observing night

Each location will have different observing night parameters

```
my_time_grid_one = time_grid_from_range(observe_night_one, time_resolution = 0.1 * u.h)
```

```
fig, ax = plt.subplots(
    figsize = (12, 6), 
    constrained_layout = True
)

for my_object in target_list:
    ax = plot_airmass(my_object, my_location_one, my_time_grid_one)

ax.legend(loc=0,shadow=True);
```

```
my_time_grid_two = time_grid_from_range(observe_night_two, time_resolution = 0.1 * u.h)
```

```
fig, ax = plt.subplots(
    figsize = (12, 6), 
    constrained_layout = True
)

for my_object in target_list:
    ax = plot_airmass(my_object, my_location_two, my_time_grid_two)

ax.legend(loc=0,shadow=True);
```

```
my_time_grid_three = time_grid_from_range(observe_night_three, time_resolution = 0.1 * u.h)
```

```
fig, ax = plt.subplots(
    figsize = (12, 6), 
    constrained_layout = True
)

for my_object in target_list:
    ax = plot_airmass(my_object, my_location_three, my_time_grid_three)

ax.legend(loc=0,shadow=True);
```
