The range of a projectile launched with a velocity (v) at and angle ($\theta$) is

$$ \large
R\ =\  {v^2 \over g}\ sin(2\theta)
$$

Problem 1
- Find R for v = 123 mph (miles per hour) and $\theta$ = 1000 arc minutes

```
# Write a function

def find_1range (my_1velocity, my_1angle):
    result=(((my_1velocity)**2)/const.g0) * np.sin(2*my_1angle)
    return result.decompose()
```

```
# Run the Function
my_1range=find_1range(
    my_1velocity=123*(imperial.mi / u.h),
    my_1angle=1000*(u.arcmin)
)

print(my_1range)
```

```
# Format the output

my_1format=f"The range would be {my_1range.value:.2f}m"

print(my_1format)
```

Problem 3

surface
of the Earth.
Expess your answer in minutes
(make sure your answer makes sense!).

```
# Write a function

def find_3period ( my_3distance):
    result=(((4*((np.pi)**2))/(const.G * const.M_earth))*(my_3distance+const.R_earth)**3)**(1/2)
    return result.decompose()
```

Problem 5

500 miles
has a duration of 3 minutes and 33 seconds. Calculate at what altitude, above the Earth's surface, you would have to orbit to go 1000 miles in this time.

.

```
# Write a function

def find_5altitude(my_5time, my_5distance):
    my_5v= my_5distance/my_5time
    my_5r = (const.G * const.M_earth)/ (my_5v)**2
    my_altitude = my_5r - const.R_earth

    return my_altitude.decompose()
```

Earth?

The self gravitational potential energy of the Earth is:
$$ \large
PE \ = \ \frac{3}{5} \cdot \frac{GM_{\oplus}^2}{R_{\oplus}}
$$
This is the amount of energy you need to give the Earth to move all of its components pieces infinitely far away (i.e. to destroy it!).

Problem 12
- Calculate how many Moon-mass objects (Moon mass = 0.0123 $M_{\oplus}$) you whould have to hit the Earth with at 45,000 mph to destroy the Earth.

```
# Write a function

def find_12number(my_12speed):
    PE_earth = (3/5) * ((const.G * (const.M_earth**2)) / const.R_earth)
    m_moon = 0.0123 * const.M_earth
    KE_moon = 0.5 * m_moon * (my_12speed**2)
    result = (PE_earth / KE_moon)

    return result.decompose()
```

```
# Run the Function

my_12speed = (45000 * imperial.mi / u.h)
my_12number = find_12number(my_12speed)

print(my_12number)
```

```
# Format the output

my_12format=f"The range would be {my_12number.value:.2f}"

print(my_12format)
```

Under Pressure

For objects in hydrostatic equilibrium, like planets and stars, the pressure at a distance (
r
) from the
core
of an object of radius (
R
) and density ($\rho$) is:
$$ \large
P(r) = \frac{2}{3} \pi G \rho^{2} \left( R^{2} - r^{2} \right)
$$

Problem 14
- Diamonds need a minimum pressure of about 5 $\times$ 10$^{9}$ Pa to form.  Calculate how
deep below the crust of the Earth
do you need to go to form diamonds. (Density of Earth = 5,520 kg/m$^{3}$).
Express your answer in km
.

```
# Write a function

def find_14depth(my_14pressure, my_14density):
    result = const.R_earth - np.sqrt(const.R_earth**2 - (3*my_14pressure)/(2 * (np.pi) * const.G * (my_14density**2)))
    return result.decompose()
```

```
# Run the Function

my_14depth = find_14depth(
    my_14pressure= 5e9 * (u.Pa),
    my_14density= 5520 * (u.kg / (u.m)**3)).to(u.km)
print(my_14depth)
```
