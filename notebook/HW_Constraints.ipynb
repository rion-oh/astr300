```
from astropy.table import QTable

from astropy.coordinates import SkyCoord
from astroplan import observability_table
from astroplan import AltitudeConstraint, MoonSeparationConstraint, SunSeparationConstraint
```

Targets

Setup the targets from the data file
https://uwashington-astro300.github.io/A300_Data/BrightStars_50.csv
and create a target list.
The columns are:
Name
RA (hours)
DEC (deg)
Add units to the columns

```
target_table = QTable.read('https://uwashington-astro300.github.io/A300_Data/BrightStars_50.csv', 
                           format='ascii.csv')
```

```
target_table['RA'].unit = u.hourangle
target_table['DEC'].unit = u.deg

target_table[0:3]
```

```
my_targets = [FixedTarget(coord = SkyCoord(ra = my_RA, dec = my_DEC), name = my_Name) #FixedTraget: 관측시에 좌표를 input하는 함수
           for my_Name, my_RA, my_DEC in target_table]
```

For each of the 3 locations create an
observability_table
with the following constraints

Use the new target list

Same observing date (12/24/2026)
The time range will be different for each location
Constraints:
Target > 6 hours from Sun
Target > 60 degrees from Moon
Target Altitude > 60 degrees
Print out the table for each location showing
ONLY
the objects that are observable

```
observing_date=Time("2026-12-24")
```

```
my_time_range_one = [observe_night_one[0],observe_night_one[1]]
my_time_range_two = [observe_night_two[0],observe_night_two[1]]
my_time_range_three = [observe_night_three[0],observe_night_three[1]]
```

```
my_constraints = [SunSeparationConstraint(6 * u.hourangle)]
my_constraints.append(MoonSeparationConstraint(60 * u.deg))
my_constraints.append(AltitudeConstraint(min=60*u.deg, max=None))
```

```
observing_table_one = observability_table(my_constraints,
                                      my_location_one,
                                      my_targets,
                                      time_range = my_time_range_one
                                     )
observing_table_one[observing_table_one['ever observable'] == True]
```

```
observing_table_two = observability_table(my_constraints,
                                      my_location_two,
                                      my_targets,
                                      time_range = my_time_range_two
                                     )
observing_table_two[observing_table_two['ever observable'] == True]
```

```
observing_table_three = observability_table(my_constraints,
                                      my_location_three,
                                      my_targets,
                                      time_range = my_time_range_three
                                     )
observing_table_three[observing_table_three['ever observable'] == True]
```
